## 第二题 理想国数据分析

- 下载理想国英文版本txt文件，网址为：http://www.gutenberg.org/cache/epub/150/pg150.txt
- 将下载文件保存到某一个文件夹。
- 使用python读取pg150.txt文件
- 对读入的文件进行清洗
- 计算每个单词出现的频次，格式：单词，频次
- 将出现次数最多的100个单词及其频次写入硬盘。




提示（非强制使用）：

- 使用with open打开
- 使用readlines读取
- 使用join合并字符串构成的列表
- 使用replace清洗
- 使用split分词
- 去除列表中的空元素''
- from collections import Counter 
- 使用Counter计数
- 对Counter的计算结果使用most_common输出最多的100个单词
- 将结果写入硬盘


```python
with open(file="D:\XA\css-master\homework\理想国.txt",mode="r",encoding="utf-8")as file:
    txt=file.readlines()
    file.close()
print(txt)
```

```python
txt=''.join(txt).strip('\n')
print(txt)
```

```python
import pandas as pd
```

```python
data = txt.replace('.', '').replace(',', '').replace('"', '').replace('(.*?)','').replace(';','').replace('*','')
print(data)
```

```python
words=data.split()
from collections import Counter
word_counts = Counter(words)
top_100 = word_counts.most_common(100)
print(top_100)
```
